# 1. Distribution

Highly recommend to play with the bandwidth argument in order not to miss a specific pattern in the data

- Input: A vector of numbers. It can be a list, or the column of a data frame. It can also show multiple variables.
- Popular choices: density plot, histogram, violin, boxplot

## 1.1 Kernel Density Plot

### One Variable

```python
df = sns.load_dataset('iris')
 
# Make default density plot
sns.kdeplot(df['sepal_length'], shade=True, bw=0.2, color='y')
plt.show()
```

### Multiple Variables

```python
# plot two variables
# plot of 2 variables
p1=sns.kdeplot(df['sepal_width'], shade=True, color="r", bw=0.2)
p1=sns.kdeplot(df['sepal_length'], shade=True, color="b", bw=0.2)
plt.show()
```

## 1.2 Histogram

### One Variable

```python
# Make default histogram
sns.distplot(a=df["sepal_length"], hist=True, kde=False, rug=False)
plt.show()
```

```python
# Hist + Rug + kernel density
sns.distplot( a=df["sepal_length"], hist=True, kde=True, rug=True )
plt.show()
```

```python
# To change parameters of rug
sns.distplot( a=df["sepal_length"], rug=True,
rug_kws={"color": "r", "alpha":0.3, "linewidth": 2, "height":0.2 }
)
plt.show()
```

```python
# To change parameters of density distribution
sns.distplot( a=df["sepal_length"], kde=True,
kde_kws={"color": "g", "alpha":0.3, "linewidth": 3, "shade":True })
plt.show()
```

### Multiple Variables

```python
p2 = sns.distplot(a=df['sepal_length'], kde=True, color='y', label='Sepal Length')
p2 = sns.distplot(a=df['sepal_width'], kde=True, color='b', label='Sepal Width')
p2.legend()
plt.show()
```

```python
# plot
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
plt.show()
```

## 1.3 Boxplot

### One Variable

```python
# Make boxplot for one group only
sns.boxplot( y=df["sepal_length"] )
plt.show()
```

### Multiple Variables

```python
# Here we study the sepal length of 3 species
sns.boxplot( x=df["species"], y=df["sepal_length"] )
plt.show()
```

### Grouped Boxplot

```python
df_tips = sns.load_dataset('tips')
```

```python
# Grouped boxplot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=df_tips, palette="Set1")
ax.set_title('Total Bill Comparison Between Smoker and Non Smoker')
plt.show()
```

## 2.1 Scatter Plot

```python
# Custom the inside plot: options are: scatter | reg | resid | kde | hex
sns.jointplot(x="sepal_length", y="sepal_width", kind='scatter', data=df)
# sns.jointplot(x=df['sepal_length'], y=df['sepal_width'], kind='scatter')
plt.show()
```

### Assign different color for points from different categories

```python
df = sns.load_dataset('iris')

# Provide a dictionary to the palette argument
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=df, fit_reg=False, hue='species', legend=False,
          palette=dict(setosa="#FF0003", virginica="#0300FF", versicolor="#1FFF00"))

# Move the legend to an empty part of the plot
plt.legend(loc='lower right')

plt.show()
```

## 2.2 Heatmap

```python
# Data set
url = 'https://python-graph-gallery.com/wp-content/uploads/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
df.head()
```

```python
# Prepare a vector of color mapped to the 'cyl' column
my_palette = dict(zip(df.cyl.unique(), ["orange","yellow","brown"]))
row_colors = df.cyl.map(my_palette)
```

```python
# plot
sns.clustermap(df, metric="correlation", method="single", cmap="Blues", standard_scale=1, row_colors=row_colors)
plt.show()
```
